# 4. Linear Regression Forecast: Predict future MaxTemp trends
# Average yearly max temperature

```python

yearly_avg_temp = df.groupby('Year')['MaxTemp'].mean().dropna().reset_index()
```

# Prepare data for regression

```python

X = yearly_avg_temp['Year'].values.reshape(-1, 1)
y = yearly_avg_temp['MaxTemp'].values
model = LinearRegression()
model.fit(X, y)
```

# Predict for next 10 years

```python

future_years = np.arange(X[-1][0] + 1, X[-1][0] + 11).reshape(-1, 1)
future_preds = model.predict(future_years)
```

# Plot

```python
# Plot
plt.figure(figsize=(12, 6))
plt.plot(X, y, 'bo-', label='Historical Avg MaxTemp')
plt.plot(future_years, future_preds, 'r--', label='Forecasted MaxTemp')
plt.title('Max Temperature Trend Forecast')
plt.xlabel('Year')
plt.ylabel('Max Temperature (Â°C)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
